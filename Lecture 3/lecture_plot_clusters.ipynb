```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create two clouds of points in 2-dimensional space and plot it, divided by color
x1 = np.random.normal(0, 1, (2, 100))
x2 = np.random.normal(3, 0.1, (2, 100))

y1 = np.zeros((100, ))
y2 = np.ones((100, ))
```

```python
# Join them togheter
X = np.concatenate((x1, x2), axis=1)
Y = np.concatenate((y1, y2))
X.shape, Y.shape
```

```python
# Visualize
plt.scatter(X[0, :], X[1, :], c=Y) # c is the color
plt.show()
```
